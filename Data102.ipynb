# Box plot

Box plot มีจุดมุ่งหมายเป็นารพอตกราฟ แนวโน้ม 

```
from matplotlib import pyplot as plt 
```

```
plt.boxplot(data['x'])
```

ใช้คำสั่ง vert เพื่อในกล่องอยู่ในแนวนอน

```
plt.boxplot(data['x'],vert = False )
```

เปลี่ยน whisker เป็น 1.2 ให้มี ค่า out lier ให้มีมากกว่าเดิม

```
plt.boxplot(data['x'],vert = False, whis=1.2 )
```

ต้องการโชว์แค่ผลลัพธ์ โดยไม่โชว์ output

```
oput = plt.boxplot(data['x'],vert = False, whis=1.2 )
```

ให้แสดงค่า outlier ถ้าอยากได้ค่าของอะไรให้ใส่ใน [ ] ตามด้วย .get_xdata() โดยที่จะเอาค่า x หรือ y ก็ได้ ที่แสดงออกมาคือค่าที่เกิน

```
oput['fliers'][0].get_xdata()
```

เส้น whisker วาดจากค่า Q1(=44.1026) ถึง 1.2IQR(=22.3077) (1.2 เรากำหนดใหม่ ปกติเป็น 1.5)

```
oput['whiskers'][0].get_xdata()
```

## Bar chart

barchart คือกราฟแท่ง
x คือจุดตั้งต่ำแหน่งของแกนกราฟ
y คือความสูงของแกน มีความสูงเท่าไหร่

```
#เรียกใช้ฟังก์ชันที่จำเป็น
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```
labels = ['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27] #กำหนดตัวแปร
women_means = [25,32,34,20,25]
```

```
x = np.arange(len(labels)) # the label locations
width = 0.35 # the width of the bars
```

ax.set_ylabel('...') คือการตั้งชื่อแกน
ax.set_title('....') คือการตั้งชื่อ ของกราฟทั้งหมด
ax.set_xticks(x) ax.set_xticklabels(labels) เป็นโค้ดที่มาด้วยกันเพื่อให้เป็นค่าที่เราต้องการ
ax.legend() เป็นการติดชื่อให้กับกราฟตามที่เรากำหนดใน .bar

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='men') # + - คือให้ข้อมูลเลื่อนไปซ้ายครึ่งนึง ขวาครึ่งนึง 
rects1 = ax.bar(x + width/2, women_means, width, label='women')

ax.set_ylabel('Scores') #ตั้งชื่อให้แกน Y 
ax.set_title('Scores by group and gender') #ตั้งชื่อ กราฟ

#เปลี่ยนค่าแกน x ให้เป็นตาม lebels ที่เรากำหนดไว้ ต้องมาด้วยกันทั้ง2บรรทัด
ax.set_xticks(x)
ax.set_xticklabels(labels) #ใส่; เพื่อบอกว่าจบแล้วไม่ต้องปริ้นอะไร
ax.legend() ; #เพิ่มคำอธิบายแต่ละเส้นของกราฟ
```

การพอร์ตกราฟแท่งง่ายๆ 
โดยจะไม่สามารถกำหนดค่าแกน x ได้

```
plt.bar(x - width/2, men_means, width, label='men') #คำสั่ง.bar สามารถสร้างกราฟแท่งได้แต่ไม่สามารถเปลี่ยนแปลงค่าแกนxได้
plt.bar(x + width/2, women_means, width, label='women')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend() ;
```
